## **23. How does the presence of a discount impact the purchase amount for different categories?**

```python
# Separate data into purchases with and without discounts
discounted_data = shop[shop['Discount Applied'] == 'Yes']
non_discounted_data = shop[shop['Discount Applied'] == 'No']

# Calculate the average purchase amount by category for discounted and non-discounted purchases
avg_purchase_discounted = discounted_data.groupby('Category')['Purchase Amount (USD)'].mean()
avg_purchase_non_discounted = non_discounted_data.groupby('Category')['Purchase Amount (USD)'].mean()

# Combine the results for comparison
discount_impact = pd.DataFrame({
    'Discounted': avg_purchase_discounted,
    'Non-Discounted': avg_purchase_non_discounted
}).reset_index()
discount_impact
```

```python
# Aggregate data for total purchase amounts
total_discounted = discounted_data.groupby('Category')['Purchase Amount (USD)'].sum()
total_non_discounted = non_discounted_data.groupby('Category')['Purchase Amount (USD)'].sum()

# Combine data into a DataFrame
stacked_data = pd.DataFrame({
    'Discounted': total_discounted,
    'Non-Discounted': total_non_discounted
}).reset_index()

# Create stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(stacked_data['Category'], stacked_data['Discounted'], label='Discounted', color='skyblue')
ax.bar(stacked_data['Category'], stacked_data['Non-Discounted'], bottom=stacked_data['Discounted'], label='Non-Discounted', color='salmon')

# Add labels and title
ax.set_xlabel('Category', fontsize=12)
ax.set_ylabel('Total Purchase Amount (USD)', fontsize=12)
ax.set_title('Total Purchases by Category (Discounted vs Non-Discounted)', fontsize=14)
ax.legend()

# Display chart
plt.tight_layout()
plt.show()
```
